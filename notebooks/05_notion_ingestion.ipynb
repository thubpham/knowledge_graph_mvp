# Notion Ingestion
Fetch qualifying Notion pages, ingest them into FalkorDB, and run test queries.

```python
import sys
sys.path.insert(0, '..')

from core.graph import KnowledgeGarden
from llm_clients import LLMClient
from data.sources.notion_ingester import ingest_notion_pages

kg = KnowledgeGarden()
client = LLMClient()
```

```python
# Cell 2: Fetch and ingest all qualifying Notion pages
result = ingest_notion_pages(kg, client)
print(result)
```

```python
# Cell 3: Counts
nodes = kg.get_all_nodes()
edges = kg.get_all_edges()
print(f'Nodes : {len(nodes)}')
print(f'Edges : {len(edges)}')
print('Sample nodes:')
for n in nodes[:5]:
    print(f'  {n.id} ({n.type}) — {n.name}')
```

```python
# Cell 4: Test queries
from datetime import datetime
from retrieval.query import query

test_questions = [
    "What projects is Alice working on?",
    "Who does Alice report to?",
    "What teams are involved in the infra project?",
]

now = datetime.now()
for q in test_questions:
    print(f'Q: {q}')
    result = query(kg, q, client, now)
    if isinstance(result, list):
        for edge in result[:3]:
            print(f'  {edge.source} --[{edge.relation}]--> {edge.target}: {edge.fact}')
    else:
        print(f'  {result}')
    print()
```
